```python
UnmetDemand_stations_list = np.sort(raw_df['start station'].unique())
print(UnmetDemand_stations_list)
```

```python
start_datetime = dt.datetime.strptime("2021-09-24 00:00:00","%Y-%m-%d %H:%M:%S")
end_datetime = dt.datetime.strptime("2021-12-23 19:00:00","%Y-%m-%d %H:%M:%S")

date_range = pd.date_range(start=start_datetime, end=end_datetime, freq='H')
UnmetDemand_df = pd.DataFrame(0,columns=UnmetDemand_stations_list,index=date_range)
UnmetDemand_df
```
