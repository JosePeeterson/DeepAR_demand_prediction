```python
import json
import pandas as pd
import os
from datetime import datetime


os.chdir('/home/optimusprime/Desktop/peeterson/github/DeepAR_demand_prediction/2_freq_nbinom_LSTM/1_cluster_demand_prediction/data/results_data')

#f = open('Singapore_fewer_clstrs_all_data.json')
f = open('Singapore_fewer_clstrs_hist_avg_fut_inf_all_data.json')
data = json.load(f)
print(data.keys())
print('actuals_dem clusters: ', data['actuals_dem'].keys())
```

```python

""" actual demand/outflow"""
col_list = list(data['actuals_dem'].keys())
clusters_actual = data['actuals_dem']
start_time = "2021-10-08 16:00:00"
datetime = pd.to_datetime(start_time) + pd.to_timedelta(range(len(clusters_actual[col_list[0]])), 'h')


all_actual_df = pd.DataFrame(columns=col_list,index=datetime)
#all_actual_df.insert(0,'datetime',datetime)

for c in col_list:
    all_actual_df[c] = data['actuals_dem'][c]

# all_actual_df.index = all_actual_df.index.strftime("%d/%m/%Y %H:%M:%S")
# all_actual_df.to_csv('singapore_fewer_clstrs_all_data_actual_outflow.csv')


""" DeepAR predicted demand"""
col_list = list(data['all_clstrs_deepar_pred'].keys())
clusters_deepar_pred = data['all_clstrs_deepar_pred']
start_time = "2021-10-08 16:00:00"
datetime = pd.to_datetime(start_time) + pd.to_timedelta(range(len(clusters_deepar_pred[col_list[0]])), 'h')


all_deepar_pred_df = pd.DataFrame(columns=col_list,index=datetime)
#all_deepar_pred_df.insert(0,'datetime',datetime)

for c in col_list:
    all_deepar_pred_df[c] = data['all_clstrs_deepar_pred'][c]

# all_deepar_pred_df.index = all_deepar_pred_df.index.strftime("%d/%m/%Y %H:%M:%S")
# all_deepar_pred_df.to_csv('singapore_fewer_clstrs_all_data_deepar_pred_outflow.csv')
all_deepar_pred_df
```
