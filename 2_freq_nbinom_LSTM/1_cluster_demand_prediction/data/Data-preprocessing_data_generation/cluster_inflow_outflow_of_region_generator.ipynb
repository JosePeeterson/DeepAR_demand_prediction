```python
import pandas as pd
import numpy as np
import os
```

```python
"""
IDentify clusters of this region
"""

region_name = "Tampines"

os.chdir('/home/optimusprime/Desktop/peeterson/github/DeepAR_demand_prediction/2_freq_nbinom_LSTM/1_cluster_demand_prediction/data/demand_data/standalone')
stn_df = pd.read_csv("station_cc_id_60_0.5.csv")
region_df = stn_df[stn_df["name"].apply( lambda x :  x.split('/')[0] ) == region_name]
cluster_names = np.sort(region_df["cc_id"].unique())
cluster_names = cluster_names.astype(str)
cluster_names
```

```python
inf_df = pd.read_csv("inflow_station_dem.csv")
inf_df = inf_df.loc[:,cluster_names]

outf_df = pd.read_csv("outflow_station_dem.csv")
outf_df = outf_df.loc[:,cluster_names]
```
